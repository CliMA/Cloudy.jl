# Choosing sigma given Âµ

```julia
using Cloudy.BasisFunctions
using Cloudy.Collocation
```

```julia
using Plots
```

By trial and error (visual plotting); for different numbers of rbf

```julia
Nb = 3
xmin = 10
xmax = 150
rbf_mu = exp.(range(log(xmin), stop=log(xmax), length=Nb) |> collect)
rbf_pts = append!([0.0], rbf_mu)
c = ones(Nb)
xplot = collect(range(0.0, stop=250.0, length=1000))

smooth_factor = 1.3
rbf_sigma = (rbf_pts[2:end]-rbf_pts[1:end-1])/smooth_factor
basis = Array{PrimitiveUnivariateBasisFunc}(undef, Nb)
for i = 1:Nb
    if (rbf_mu[i] - rbf_sigma[i]*1.6) < 0.0
        rbf_sigma[i] = rbf_mu[i]/1.6
    end
    if i > 2
        if (rbf_mu[i] - rbf_sigma[i]*1.6) < rbf_mu[i-2]
            rbf_sigma[i] = (rbf_mu[i] - rbf_mu[i-2])/1.6
        end
    end
    basis[i] = GaussianBasisFunction(rbf_mu[i], rbf_sigma[i])
    println(basis[i])
end
plot(xplot, 
    evaluate_rbf(basis, c, xplot),
    lw=2,
    label="1.3")
```
